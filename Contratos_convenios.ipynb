```python
%pip install requests
import requests

# Definindo a URL da API
url_convenios = 'https://api-dados-abertos.cearatransparente.ce.gov.br/transparencia/contratos/convenios?page=133&data_assinatura_inicio=01%2F01%2F2024&data_assinatura_fim=31%2F12%2F2025'
# Fazendo a chamada da API
response = requests.get(url_convenios, headers={'accept': 'application/json'})
# Verificando se a requisição foi bem sucedida
if response.status_code == 200:
	# Carregando os dados da resposta em um DataFrame
	data = response.json()
	print(data)
else:
	print(f"Erro na requisição: {response.status_code}")

url_contratos = 'https://api-dados-abertos.cearatransparente.ce.gov.br/transparencia/contratos/contratos?page=444&data_assinatura_inicio=01%2F01%2F2024&data_assinatura_fim=31%2F12%2F2024'
# Fazendo a chamada da API
response = requests.get(url_contratos, headers={'accept': 'application/json'})
# Verificando se a requisição foi bem sucedida
if response.status_code == 200:
	# Carregando os dados da resposta em um DataFrame
	data = response.json()
	print(data)
else:
	print(f"Erro na requisição: {response.status_code}")
```

```python
convenios = data['data']
contratos = data['data']

# Exibe ambos os resultados
from IPython.display import display

display(convenios)
display(contratos)
```

```python
df = pd.DataFrame(convenios + contratos)
print(df)
```

```python
import pandas as pd
import numpy as np

# --- Início do Exemplo: Criando um DataFrame para demonstrar ---
# No seu código real, você já terá o seu DataFrame 'df' carregado.
data = {
    'cod_orgao': ['A01', 'B02', np.nan, 'D04'],
    'nome_contratado': ['Empresa X', 'Empresa Y', 'Empresa Z', np.nan],
    'valor_contrato': [1500.50, np.nan, 3200.00, 4100.75],
    'coluna_extra': [1, 2, 3, 4]
}
df = pd.DataFrame(data)
# --- Fim do Exemplo ---


# Lista de colunas que você quer ajustar
cols = [
    'cod_orgao',
    'cod_financiador',
    'nome_fiscal',
    'cod_gestora',
    'descricao_modalidade',
    'descricao_tipo',
    'valor_contrato',
    'calculated_valor_pago'
]

# 1. Filtra a lista para conter apenas as colunas que realmente existem no DataFrame
existing_cols = [col for col in cols if col in df.columns]

# 2. Itera sobre as colunas existentes e substitui os valores NaN por "Não Informado"
for col in existing_cols:
    df[col] = df[col].fillna('Não Informado')

# 3. Agora você pode verificar o resultado. O comando value_counts() geralmente
#    é aplicado a uma única coluna (ou a uma série de colunas) para ver a contagem.
#    Por exemplo, para verificar a coluna 'nome_contratado':
print(df['nome_contratado'].value_counts(dropna=False))

print("\nDataFrame resultante:")
print(df)
```
